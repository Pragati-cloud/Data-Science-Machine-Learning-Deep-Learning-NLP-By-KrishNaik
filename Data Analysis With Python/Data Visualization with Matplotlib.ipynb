```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[3,6,9,12,15,18]

plt.plot(x,y)
```

```python
x=[1,2,3,4,5,6]
y=[3,23,9,15,2,17]

plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y asix')
plt.title('Line Plot')
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[3,23,9,15,2,17]

plt.plot(x,y,color='purple', linestyle='--', marker='o') #linewidth and markersize
plt.grid(True)
plt.xlabel('X axis')
plt.ylabel('Y asix')
plt.title('Line Plot')
plt.show()
```

```python
#Mutiple plots

x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2 = [2,4,6,8,10]

plt.figure(figsize=(9,5))

plt.subplot(1,2,1) # row , column , plot
plt.plot(x,y1,color='green')
plt.title('Plot 1')

plt.subplot(1,2,2)
plt.plot(y1,x,color='green')
plt.title('Plot 1')
```

```python
# bar plot

Categories = ['A','B','C','D','E']
values = [3,4,2,7,1]

plt.bar(Categories,values,color='pink')
```

```python
# Histogram

data = [1,1,2,4,4,4,6,6,6,5,5,4,7,7,7]

plt.hist(data, bins=6, color='orange')
```

```python
# scatter plot

x=[2,4,5,6,7,8]
y=[6,4,3,6,3,7]

plt.scatter(x,y, marker='*')
```

```python
## pie chart

labels = ['A','R','F','X','G']
sizes=[34,23,66,75,36]

colors =['gold','yellowgreen','lightcoral','lightskyblue','silver']
explode =(0.2,0,0,0,0)

plt.pie(sizes,labels=labels, explode=explode , shadow=True, colors=colors, autopct='%1.1f%%')
```
