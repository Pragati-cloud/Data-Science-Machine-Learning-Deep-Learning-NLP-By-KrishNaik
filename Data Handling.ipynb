Nominal / onehotencoding
Technique used to represent categorical data as numerical data , suitable for ml algo . Each category is represented as binary vector where each bit is corresponding to new category.

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
##create dataframe
df = pd.DataFrame({'color':['red','blue','green','green','red','blue']})
```

```python
df.head()
```

```python
#instance of onehotencoder
encoder= OneHotEncoder()
```

```python
encoder = encoder.fit_transform(df[['color']]).toarray()
```
